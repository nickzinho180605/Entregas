```
# ==========================================
# SISTEMA DE GESTIÓN DE BIBLIOTECA
# Autor: Proyecto Académico
# Lenguaje: Python
# Algoritmo: Búsqueda Binaria
# ==========================================

biblioteca = []


def ordenar_libros():
    biblioteca.sort(key=lambda libro: libro["codigo"])


def registrar_libro():
    print("\n--- REGISTRAR LIBRO ---")

    codigo = int(input("Código del libro: "))

    for libro in biblioteca:
        if libro["codigo"] == codigo:
            print("Error: El código ya existe.")
            return

    titulo = input("Título: ")
    autor = input("Autor: ")
    anio = input("Año de publicación: ")

    biblioteca.append({
        "codigo": codigo,
        "titulo": titulo,
        "autor": autor,
        "anio": anio
    })

    ordenar_libros()

    print("Libro registrado correctamente.")


def busqueda_binaria(codigo):
    ordenar_libros()

    izquierda = 0
    derecha = len(biblioteca) - 1

    while izquierda <= derecha:
        medio = (izquierda + derecha) // 2

        if biblioteca[medio]["codigo"] == codigo:
            return medio

        elif biblioteca[medio]["codigo"] < codigo:
            izquierda = medio + 1

        else:
            derecha = medio - 1

    return -1


def buscar_libro():
    print("\n--- BUSCAR LIBRO ---")

    codigo = int(input("Ingrese código a buscar: "))

    posicion = busqueda_binaria(codigo)

    if posicion != -1:
        libro = biblioteca[posicion]

        print("\nLibro encontrado")
        print("Código:", libro["codigo"])
        print("Título:", libro["titulo"])
        print("Autor:", libro["autor"])
        print("Año:", libro["anio"])

    else:
        print("Libro no encontrado.")


def mostrar_libros():
    print("\n--- LISTA DE LIBROS ---")

    if len(biblioteca) == 0:
        print("No existen libros registrados.")
        return

    print("-" * 70)

    for libro in biblioteca:
        print(
            f"Código: {libro['codigo']} | "
            f"Título: {libro['titulo']} | "
            f"Autor: {libro['autor']} | "
            f"Año: {libro['anio']}"
        )

    print("-" * 70)


def actualizar_libro():
    print("\n--- ACTUALIZAR LIBRO ---")

    codigo = int(input("Ingrese código del libro: "))

    posicion = busqueda_binaria(codigo)

    if posicion == -1:
        print("Libro no encontrado.")
        return

    print("Ingrese los nuevos datos")

    biblioteca[posicion]["titulo"] = input("Nuevo título: ")
    biblioteca[posicion]["autor"] = input("Nuevo autor: ")
    biblioteca[posicion]["anio"] = input("Nuevo año: ")

    print("Libro actualizado correctamente.")


def eliminar_libro():
    print("\n--- ELIMINAR LIBRO ---")

    codigo = int(input("Ingrese código del libro: "))

    posicion = busqueda_binaria(codigo)

    if posicion == -1:
        print("Libro no encontrado.")
        return

    del biblioteca[posicion]

    print("Libro eliminado correctamente.")

# Menú principal
while True:

    print("\n================================")
    print(" SISTEMA DE GESTIÓN DE BIBLIOTECA ")
    print("================================")
    print("1. Registrar libro")
    print("2. Buscar libro")
    print("3. Mostrar libros")
    print("4. Actualizar libro")
    print("5. Eliminar libro")
    print("6. Salir")

    opcion = input("Seleccione una opción: ")

    if opcion == "1":
        registrar_libro()

    elif opcion == "2":
        buscar_libro()

    elif opcion == "3":
        mostrar_libros()

    elif opcion == "4":
        actualizar_libro()

    elif opcion == "5":
        eliminar_libro()

    elif opcion == "6":
        print("Gracias por utilizar el sistema.")
        break

    else:
        print("Opción inválida.")
```
